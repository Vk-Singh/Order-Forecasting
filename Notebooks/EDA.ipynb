```python
def stats(data, warehouses):
    temp_dict = {}
    for warehouse in warehouses:
        temp_dict[warehouse] = data[data['warehouse'] == warehouse]['orders'].describe()
        #print(f' Details for Orders for warehouse {warehouse} | \n {temp_dict[warehouse]}')
    
    return temp_dict
```
